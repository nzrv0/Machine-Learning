### Data egnineering

```python
# converting currencies
df["currency_x"].value_counts()
```

```python
# converting all currencies to azn
dollar = df.loc[df["currency_x"] == "$", "price_x"]
euro = df.loc[df["currency_x"] == "€", "price_x"]

df.loc[df["currency_x"] == "$", "price_x"] = dollar * 1.7
df.loc[df["currency_x"] == "€", "price_x"] = euro * 1.8
```

```python
# there is only kimoterage unit is NA is where kiolemtrage number is 0 so there will be no change we're just not use this column
df.loc[df["kilometrage_unit"].isna(), "kilometrage_num"].value_counts()
```

```python
# some engine displacment numbers are 0
df.loc[df["engine_displacement_unit"].isna(), "engine_displacement_num"]
```

```python
# only electros has no Litr paramter
df["Mühərrik"].str.split("/", expand=True).value_counts()
```

```python
# selecting every a.g values
engine_parsed = df["Mühərrik"].str.split("/", expand=True)

engine_parsed_ag = pd.DataFrame(np.where((engine_parsed.iloc[::, -2] == " Elektro"), engine_parsed[0], engine_parsed[1]), index=df["Mühərrik"].index)
df["Hourse Power"] = engine_parsed_ag[0].str.split("a.g.", expand=True)[0]

# some values of hourse power is NaN due to these are the motorcycles 
df.loc[df["Hourse Power"].isna(), "Mühərrik"]
```

```python
engine_fuel_type = np.where((engine_parsed.iloc[::, -2] == " Elektro"), engine_parsed[1], engine_parsed[2])

engine_fuel_type = [i.split(" ")[-1] for i in engine_fuel_type]

df["Fuel Type"] = pd.DataFrame(engine_fuel_type, index=df["Mühərrik"].index)
```

```python
df = df.dropna(subset=["Hourse Power"])
```

```python
df["Hourse Power"] = df["Hourse Power"].astype(int)
```

```python
# Some operations on owners column
df.loc[df["Sahiblər"]=="4 və daha çox", "Sahiblər"] = 4
df.loc[df["Sahiblər"]==0, "Sahiblər"] = 1
df["Sahiblər"] = df["Sahiblər"].fillna(df["Sahiblər"].median())
df["Sahiblər"] = df["Sahiblər"].astype(int)
```

```python
df["barter"] = df["barter"].fillna("Barter mümkün deyil")
df["loan"] = df["loan"].fillna("Kreditdə deyil")
df["Qəzalı"] = df["Qəzalı"].fillna("Xeyr")
```

```python
df.loc[df["Yerlərin sayı"]=="8+", "Yerlərin sayı"] = 8
df["Yerlərin sayı"] = df["Yerlərin sayı"].fillna(df["Yerlərin sayı"].median())
df["Yerlərin sayı"] = df["Yerlərin sayı"].astype("int")
```

```python
df["production_year"].min()
```

```python
min_production_year = df.loc[df["production_year"] != 0, "production_year"].min()
df.loc[df["production_year"] == 0, "production_year"] = min_production_year
```

```python
categorical_values = ["barter", "loan", "Qəzalı", "Vəziyyəti", "Yeni", "Sürətlər qutusu", "Ötürücü", "Fuel Type", "Rəng"]

_, axes = plt.subplots(3, 3, figsize=(16, 12))
axes = axes.ravel()
for index, item in enumerate(categorical_values):
    category = df[categorical_values[index]].value_counts()
    axes[index].barh(category.index, category.values)
    axes[index].set_title(item)
```
